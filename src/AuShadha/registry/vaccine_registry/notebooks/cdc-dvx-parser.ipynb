```python
#! /usr/bin/python

##################################################
# Project: This code is part of AuShadha Project
# Author: Dr. Easwar T.R
# Date: 09 Aug 2014
# License: GNU-GPl Version 3 
##################################################

"""
  Parser for the DVX, DVX-VIS files from CDC.
  DVX File contains the data for all the CDC approved Vaccines.
  The XML is available from CDC website. http://cdc.gov
  For usage details, license please consult the website
"""


#Imports
import os
import sys
import copy
import json
from xml.etree import cElementTree as ET


#Constants
ROOT_PATH = os.path.abspath(os.path.curdir)
APP_PATH = os.path.abspath(os.path.pardir)
DATA_DIR = ROOT_PATH + '/data/cdc/'
APP_FIXTURE_PATH = APP_PATH + '/fixtures/initial_data.json'
DVX_XML = ET.parse(DATA_DIR+"/cvx.xml") 
DVX_ROOT = DVX_XML.getroot()

#Basic parsed info
print("Parsed the DVX XML")
print("ROOT tag: " + DVX_ROOT.tag)

#Store vaccine info as iterable to be used later
all_vaccines = (lambda: [v for v in DVX_ROOT.findall('CVXInfo')])()            # Store all the Vaccines in a iterable form
vaccine_tag_names = (lambda: [t.tag for t in all_vaccines[0].getchildren()])() # Just to check for outliers

def test_presence_of_cvx_codes():
    """ Simple test to see if CVXCode is present in all XMl vaccine entries"""
    count = 0
    for t in all_vaccines:
        for x in t.findall('CVXCode'):
            if x.text == None:
                count += 1
    return count

def build_data():
    """ Builds the data from the XML to make it ready for output """
    data = []
    for _v in all_vaccines:
        data_to_append = {}
        for elem in _v.getchildren():
            data_to_append[elem.tag] = elem.text
        data.append(data_to_append) 
    return data

def to_json():
    """ Returns the JSON of the Vaccine CVX Data """
    return json.dumps(build_data(),indent=4)



def to_django_fixture(format = 'json'):
    
    """ Method to generate a Django Fixture. By default generates JSON fixture"""
    
    data = []
    vaccine_detail_model_def = {'pk': None, 
                                 'fields': {'vaccine_id':""},
                                 'model': 'vaccine_registry.VaccineDetail'
                                }
    
    vaccine_data_model_def = {'pk': None, 
                             'fields': {'field_name': '',
                                        'field_value': '',
                                        'vaccine_fk': ''
                                        },
                             'model': 'vaccine_registry.VaccineData'
                             }
    vaccine_data_pk=1    
    for _i in build_data():
        _v_detail = copy.deepcopy(vaccine_detail_model_def)
        _v_detail['pk']= int(_i['CVXCode'])
        _v_detail['fields']['vaccine_id']= int(_i['CVXCode'])
        data.append(_v_detail)

        for k,v in _i.iteritems():
            _v_data = copy.deepcopy(vaccine_data_model_def)
            _v_data['pk']= vaccine_data_pk
            _v_data['fields']['vaccine_fk']= int(_i['CVXCode'])
            _v_data['fields']['field_name'] = k
            _v_data['fields']['field_value'] = v
            data.append(_v_data)
            vaccine_data_pk += 1
    with open('initial_cvx_fixture_data.json','wb') as out_file:
        out_file.write(json.dumps(data,indent=4))
        out_file.close()
    with open(APP_FIXTURE_PATH,'a') as out_file:
        out_file.write(json.dumps(data,indent=4))    
        out_file.close()
    print(
    '''
      Fixture file Written.
      This should be moved to 'fixtures' directory
      For extra-compression please use gzip on the fixture file
    '''
    )
    return json.dumps(data,indent=4)
```

```python
#Creates a JSON output and generates a fixture
#If there is a fixture that is already generated this WILL overwrite it
print(to_django_fixture())
```
