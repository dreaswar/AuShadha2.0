```python
#! /usr/bin/python

##################################################
# Project: This code is part of AuShadha Project
# Author: Dr. Easwar T.R
# Date: 09 Aug 2014
# License: GNU-GPl Version 3 
##################################################

"""
  Parser for HTML of the Vaccination Schedule
  Parsed data of the schedule along with the footnotes can be stored in DB
"""

import bs4
import json
import os
import datetime
import copy

ROOT_PATH = os.path.abspath(os.path.curdir)
DATA_DIR = ROOT_PATH + '/data/'
CDC_DIR = DATA_DIR + '/cdc/'

CHILD_PATH = CDC_DIR + 'cdc_child_and_adolescent_immunisation_schedule_usa_compliant.html'
ADULT_PATH = CDC_DIR + 'cdc_adult_immunisation_schedule_usa_compliant.html'

get_page_text = lambda p: bs4.BeautifulSoup( open(p,'r').read())

get_all_tables = lambda p: p.select('table')

get_vaccine_name = lambda elem: elem.a.get_text().strip().replace('\n','') if elem.a else None

get_elem_attr = lambda elem,attr: elem[attr] if elem.has_key(attr) else sanitise_empty_attr_table_cell(elem,attr)

td_headers_and_title = lambda td,prefix: {'vaccine_label': get_elem_attr(td,'headers')[0],
                                          'age_range': ' '.join( get_elem_attr(td,'headers')[1:] ),
                                          'recommendation': get_elem_attr(td,'title'),
                                          'footnote': get_footnote(td,prefix),
                                          'class':  get_elem_attr(td,'class'),
                                   }

def sanitise_empty_attr_table_cell(cell,attr):
    if attr == 'headers':
        colspan = int(cell['colspan']) if cell.has_key('colspan') else 1
        vaccine_name = get_vaccine_name(cell.find_parent('tr')).replace(' ','_').replace(',','_').lower()
        headers = cell.find_parent('table').tr.find_all('th')[2:]
        h = [h['id'] for h in headers]
        h.insert(0,vaccine_name)
        return h



def get_footnote(elem,prefix):
    page = {'child': CHILD_IMMUNISATION_PAGE,
            'adult': ADULT_IMMUNISATION_PAGE
            }
    if elem and elem.a:
        q_string = 'a[name="'+ elem.a['href'].replace('#','')+'"]'
        q = page[prefix].select(q_string)
        if q:
            return q[0].find_parent('li').prettify()
        return ''
    return ''

CHILD_IMMUNISATION_PAGE = get_page_text(CHILD_PATH)
ADULT_IMMUNISATION_PAGE = get_page_text(ADULT_PATH)

all_child_tables = get_all_tables(CHILD_IMMUNISATION_PAGE)
all_adult_tables = get_all_tables(ADULT_IMMUNISATION_PAGE)

def export_vaccine_schedule_data(tables,prefix,format='json'):
    file_prefix = ROOT_PATH+'/vaccine_schedule'+prefix+'.json'
    _data = []
    for tbl in tables:
        d = {}
        d['caption'] = tbl.caption.get_text() if tbl.caption else prefix
        d['age_ranges'] = [ i.get_text().split(' ')[0] for i in tbl.tr.find_all('th')[1:]]
        d['map'] = [{'vaccine': { get_vaccine_name(i):[td_headers_and_title(td,prefix) for td in i.find_all('td')] },
                     'footnote': get_footnote(i,prefix),
                     }for i in tbl.find_all('tr')[1:] if get_vaccine_name(i)
                   ]
        _data.append(d)
    data = json.dumps(_data, indent=4)
    print("JSON data exported...")

    with open(file_prefix,'w') as out_file:
        print("JSON data being written to a file...")
        out_file.write(data)
        out_file.close()
        print("JSON data successfully written")

    return data

print( export_vaccine_schedule_data(all_child_tables[1:3],'child') )

print( export_vaccine_schedule_data(all_adult_tables[1:],'adult') )
```
